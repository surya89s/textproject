```python
import pandas as pd

# Read the log files from the system 
df1=pd.read_csv("logsOutput0.csv")
df2=pd.read_csv("logsOutput1.csv",names=df1.columns)
df=pd.concat((df1,df2),axis=0,ignore_index=True)
```

```python
troubleShoot = df['Troubleshooting_Conc']
```

```python
agent = df['AGENT_NAME']
```

```python
# Read the carefully selected vocabulary set 
#              based on curation of the list
#              that has been extracted based
#              on frequency and context usage 

# Load the computer word glossary terms 
wordbase = [line.rstrip('\n') for line in open('wordbase.txt')]

# Load the operational word glossary terms 
opwordbase = [line.rstrip('\n') for line in open('opwordbase.txt')]

# Load the tech support functional words
funcwordbase  = [line.rstrip('\n') for line in open('funcwordbase.txt')]
```

```python
# Extract agent data to find clue words from the sample
#    The data has the following variable set identifiers
#    - computer terminology
#    - dell operational terminologies
#    - tech support and troubleshoot keywords 

# Take a sample transaction log 
agentData = troubleShoot[1989]

# Tokenize the data set to extract the keywords
from nltk.tokenize import sent_tokenize, word_tokenize
agentDataStrip = word_tokenize(agentData)

# Extract the frequency map of the log transaction
#        Data set freuqncy mapping 
from collections import Counter
# 
#        computer terminology :  word frequency
#
commonWords = []
for i in range(0,len(agentDataStrip)):
    for j in range(0,len(wordbase)):
        if wordbase[j] == agentDataStrip[i].lower():
            commonWords.append(wordbase[j])
        else:
            pass
wbcounts = Counter(commonWords)   # wordbase frequency count   <--   
#    
#       dell operational terminologies : word frequency 
# 
commonWords = []
for i in range(0,len(agentDataStrip)):
    for j in range(0,len(opwordbase)):
        if opwordbase[j].lower() == agentDataStrip[i].lower():
            commonWords.append(opwordbase[j])
        else:
            pass
opcounts = Counter(commonWords)   # opwordbase frequency count  <--
#
#      tech support and troubleshoot keywords 
# 
commonWords = []
for i in range(0,len(agentDataStrip)):
    for j in range(0,len(funcwordbase)):
        if funcwordbase[j].lower() == agentDataStrip[i].lower():
            #if not in commonWords:
            commonWords.append(funcwordbase[j])
        else:
            pass
funccounts = Counter(commonWords)   # opwordbase frequency count  <--
#  Summary of all counts : stored as dictionary 
```
